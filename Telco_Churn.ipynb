```python
import pandas as pd
from tabulate import tabulate
from termcolor import colored

# Đọc dữ liệu từ file Excel trên Google Drive
file_path = "/content/drive/My Drive/Churn Prediction/dataset/a_IBM Telco Customers Churn Datasets.xlsx"
sheets = pd.ExcelFile(file_path).sheet_names
dataframes = {sheet: pd.read_excel(file_path, sheet_name=sheet) for sheet in sheets}

# Danh sách sheet không cần hiển thị
exclude_sheets = ['Population', 'Telco_Churn']

# Hàm hiển thị bảng dữ liệu đẹp hơn
def display_tabulate(df, sheet_name):
    print(colored(f"\n=== {sheet_name} ===\n", "blue", attrs=["bold"]))  # Tiêu đề màu xanh dương
    print(tabulate(df.head(5), headers="keys", tablefmt="fancy_grid", showindex=False))

# Duyệt qua từng sheet và hiển thị nếu không thuộc danh sách loại trừ
for sheet, df in dataframes.items():
    if sheet not in exclude_sheets:
        display_tabulate(df, sheet)

```
